### разобраться как вывести список репозиториев для конкретного пользователя, сохранить JSON-вывод в файле *.json.

```python
url = 'https://api.github.com/users/edivet92/repos'
r = requests.get(url=url)
```

```python
r.status_code
```

```python
r.json()
```

```python
for res in r.json():
    print(res['name'])
```

```python
repos = [rep['name'] for rep in r.json()]
```

```python
with open('repos_dump.json', 'w') as f:
    json.dump(repos, f)
```
